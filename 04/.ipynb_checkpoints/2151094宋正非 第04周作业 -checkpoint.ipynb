### 第1题 三个整数的最大公约数与最小公倍数
- 输入3个正整数，输出最大公因数与最小公倍数
  


```python
# 请编写程序， 代码写在此处
a=int(input("请输入一个正整数"))
b=int(input("请输入一个正整数"))
c=int(input("请输入一个正整数"))
import math
gys=math.gcd(a,b,c)
d=math.gcd(a,b)*math.gcd(a,c)*math.gcd(b,c)
gbs=(a*b*c)//d
print("最大公因数是",gys)
print("最小公倍数是",gbs)
```

### 第2题  数字游戏
- 有一种数字游戏，从任意正奇数开始，重复做以下运算：若为奇数，乘以3后加1；若为偶数，除以2。最后总可以得到数1。例如，奇数13运算转化的步骤为：
13->40->20->10->5->16->8->4->2->1
共进行了9步后完成。
- 请设计程序,输入一个奇数m，输出转化为1所需的转化步数，并输出转化步骤。如输入13，则输出：  
13->40->20->10->5->16->8->4->2->1   
9

```python
# 代码完成在此处
a=int(input("请输入一个正奇数"))
print(a,end="")
i=0
while a!=1:
    if a%2!=0:
        a=a*3+1
        print(" ->",a,end="")
    else:
        a=a//2
        print(" ->",a,end="")
    i+=1
print(" ,共进行了",i,"步")

```

### 第3题  页码数字
- 一本书的页数为N，页码从1开始编起，请你编程算出全部页码中，用了多少个数字X ?  
- 输入一个总的页码数N和数字X（1 ≤ N ≤ 1000，0 ≤ X ≤ 9）。
- 输出数字X的个数。
- 如：  
给定： 11    1  
输出： 4

```python
# 代码完成在此处
a=int(input("请输入一个总的页码数(1 ≤ N ≤ 1000)："))
b=int(input("请输入一个数字(0 ≤ X ≤ 9)："))
list=[]
for i in range(1,a+1):
    if i>=1 and i<=9:
        list.append(i)
    elif i>=10 and i<=99:
        ge=i%10
        shi=i//10
        c=[ge,shi]
        list.extend(c)
    elif i>=100 and i<=999:
        ge=i%10
        bai=i//100
        shi=i-ge-bai*100
        d=[ge,shi,bai]
        list.extend(d)
    elif i==1000:
        e=[1,0,0,0]
        list.extend(e)
print(list.count(b))
```

### 第4题  里程显示问题
- 有一个卡车的里程表有问题，当数字从3变到4的时候，总是跳过4变到5；当数字从7变到8的时候，直接跳到数字9。
- 即33577，下一个数字是33579；或者33399下一个数是33500。
- 如果里程表显示是15，实际上只是走了12公里。
- 问题，给定一个数（数中没有4与8），如mileage=12357,求车辆的实际里程realmailage

```python
#老师答案：1）用了个字典设置了仪表盘和八进制的对应关系2）用了列表推理每一位转换回来
#3）接着用int转换回十进制
mileage=eval(input("请输入里程"))
duiying={'0':'0','1':'1','2':'2','3':'3','5':'4','6':'5','7':'6','9':'7'}
nmileage=int("".join([duiying[each]for each in str(mileage)]),8)
print("显示{0}，实际{1}".format(mileage,nmileage))
```

```python
#代码完成在此处
m=int(input("请输入里程表显示的数值："))
r=0
for a in range(1,m+1):
    b=str(a)
    if b.count('4')==0 and b.count('8')==0:
        r+=1
    else:
        r+=0
print(r)
```
